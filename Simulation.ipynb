```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.array([
    [1, 2],
    [2, 2],
    [2, 3],
    [8, 8],
    [9, 9]
])
labels = ['A', 'B', 'C', 'D', 'E']
bandwidth = 3
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(points[:, 0], points[:, 1], c='black', s=100)

for i, label in enumerate(labels):
    ax.text(points[i][0] + 0.1, points[i][1] + 0.1, label, fontsize=12, color='blue')

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_xlabel('F1')
ax.set_ylabel('F2')
ax.set_title('Các điểm dữ liệu với nhãn')
ax.grid(True)
plt.show()
```

***Bước 2***

```python
def euclidean_distance(p1, p2):
    return np.linalg.norm(p1 - p2)

def tim_lan_can(points, center, bandwidth):
    return [i for i, pt in enumerate(points) if euclidean_distance(pt, center) < bandwidth]

#Center cuả tập điểm
def compute_mean(points):
    return np.mean(points, axis=0)

def mean_shift_step(points, labels, center_idx, bandwidth):
    
    fig, ax = plt.subplots(figsize=(8, 8))
    
    start_point = points[center_idx]
    '''
    1. Tìm điểm lân cận
    2.Đưa điểm lân cận vào tập điểm
    3. Tìm điểm trung tập của tập điểm
    '''

    neighbor_idxs = tim_lan_can(points, start_point, bandwidth)
    neighbor_points = points[neighbor_idxs]
    mean_point = compute_mean(neighbor_points)
    
    ax.scatter(points[:, 0], points[:, 1], color='black', s=100)
    for i, label in enumerate(labels):
        ax.text(points[i][0] + 0.1, points[i][1] + 0.1, label, fontsize=12)
    
    ax.scatter(*mean_point, color='red', s=100, marker='x')
    ax.text(mean_point[0] + 0.2, mean_point[1] + 0.2, 'x_new', color='red', fontsize=12)

    ax.scatter(neighbor_points[:, 0], neighbor_points[:, 1], color='blue', s=200, alpha=0.3)

    for pt in neighbor_points:
        ax.annotate('', xy=(pt[0], pt[1]), xytext=(mean_point[0], mean_point[1]),
                    arrowprops=dict(arrowstyle='->', color='blue', lw=2))
    
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.set_xlabel('F1', fontsize=14)
    ax.set_ylabel('F2', fontsize=14)
    ax.set_title('Mean Shift Bước 2', fontsize=16)
    ax.grid(True)
    plt.show()

points = np.array([
    [1, 2],  # A
    [2, 2],  # B
    [2, 3],  # C
    [8, 8],  # D
    [9, 9]   # E
])
labels = ['A', 'B', 'C', 'D', 'E']
bandwidth = 3

mean_shift_step(points, labels, center_idx=0, bandwidth=bandwidth)
```

***KẾT LUẬN***

```python
def euclidean_distance(p1, p2):
    return np.linalg.norm(p1 - p2)

def compute_mean(points):
    return np.mean(points, axis=0)

def tim_lan_can(points, center, bandwidth):
    return [i for i, pt in enumerate(points) if euclidean_distance(pt, center) < bandwidth]

def Cluster(points, labels, bandwidth):
    fig, ax = plt.subplots(figsize=(8, 8))
    
    cum1_indices = [0, 1, 2]
    cum2_indices = [3, 4]
    
    cum1 = points[cum1_indices]
    cum2 = points[cum2_indices]
    
    centroid1 = compute_mean(cum1)
    centroid2 = compute_mean(cum2)

    ax.scatter(cum1[:, 0], cum1[:, 1], color='blue', s=100, label='Cluster 1')
    ax.scatter(cum2[:, 0], cum2[:, 1], color='red', s=100, label='Cluster 2')

    for i, label in enumerate(labels):
        ax.text(points[i][0] + 0.1, points[i][1] + 0.1, label, fontsize=12)

    ax.scatter(*centroid1, color='blue', s=150, marker='x', linewidths=3)
    ax.scatter(*centroid2, color='red', s=150, marker='x', linewidths=3)

    ax.set_xlim(0, 11)
    ax.set_ylim(0, 11)
    ax.set_xlabel('F1', fontsize=14)
    ax.set_ylabel('F2', fontsize=14)
    ax.set_title('2 Cụm Dữ Liệu với Centroid', fontsize=16)
    ax.grid(True)
    ax.legend()
    plt.show()

points = np.array([
    [1, 2],  # A
    [2, 2],  # B
    [2, 3],  # C
    [8, 8],  # D
    [9, 9]   # E
])
labels = ['A', 'B', 'C', 'D', 'E']
bandwidth = 3

Cluster(points, labels, bandwidth)
```
